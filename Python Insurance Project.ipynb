```python
#4.Explore the relationship between the feature and target column using a count plot of categorical columns and a scatter plot of numerical columns

#scatter plot of numerical columns age, bmi and children:

fig,axes=plt.subplots(1,3,figsize=(21,7))

sns.scatterplot(x='age',y='charges',data=insurance,hue='smoker',ax=axes[0]) 
sns.scatterplot(x='bmi',y='charges',data=insurance,hue='smoker',ax=axes[1]) 
sns.scatterplot(x='children',y='charges',data=insurance,hue='smoker',ax=axes[2]) 
plt.show()
```

#### for 4th task we have checked the relationship between numerical veriable age , bmi and children vs charges considering the effects of smoking and non smoking croud. We found as age increases the charges for the insurance is also increased. where in the case of bmi it shows there is no direct effect of bmi to charges and for children the charges decrease as the number of childrens increase. Last but not the least we found in every situation somkers are paying more charges for insurance compare to non smokers. 

```python
### Count plot of categorical columns sex, region and smoker:

fig,axes=plt.subplots(1,3,figsize=(21,7))

sns.countplot(x='sex',data=insurance,hue='smoker',ax=axes[0]) 
sns.countplot(x='region',data=insurance,hue='smoker',ax=axes[1]) 
sns.countplot(x='smoker',data=insurance,hue='smoker',ax=axes[2]) 
plt.show()
```

#### for other part of the 4th task we have checked the relationship between categorical veriable sex, region and smokers. we will get the counts of smoking and non smoking croud. We found female has less smokers then male. futher we found the smokers in southwest are less compare to other regions and lastly we see the dataset has less smokers and more non smokers. 

```python
#5.	Perform data visualization using plots of feature vs feature:

import warnings
warnings.filterwarnings('ignore')

#created violinplot,strippolt and swarmplot for data visualization of categorical variables 
#like sex, region and smoker vs numeric variable charegs.

fig,axes=plt.subplots(1,3,figsize=(21,7)) 

sns.violinplot(x='sex',y='charges',palette='rainbow',data=insurance,ax=axes[0]) 
sns.stripplot(x='region',y='charges',data=insurance,ax=axes[1]) 
sns.swarmplot(x='smoker',y='charges',data=insurance,ax=axes[2]) 
plt.show()
```

#### for 5th task we have checked the relationship between categorical veriable sex, region and smokers vs charges. we found males and females are almost end up payinng same charges and can see same for regions as there is no out standing region paying very much more compare to other regoins. Last but not the least we found somkers are paying more charges for insurance compare to non smokers.
